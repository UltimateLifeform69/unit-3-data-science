The titanic is a cruise ship that held many people and crashed into a ice burg killing many people on board with only some passangers surviving on life boats.

```python
import pandas as pd
import matplotlib.pyplot as pip
titanic = pd.read_csv('titanic_passengers.csv')
```

## How many people on the titanic are male?


```python
sex = titanic['Sex'].value_counts()
sex.plot.pie()
```

A Majority of the passengers on the titanic were male.

## How many Passengers survived

```python
Live = titanic['Survived'].value_counts()
Live.plot.pie()
```

All of the zeros mean the people who died and all of the ones are counting who survived with a majority dying.
